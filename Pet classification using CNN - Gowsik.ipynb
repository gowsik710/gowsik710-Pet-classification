```python
# Importing the required packages 
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense,Dropout
from tensorflow.keras.optimizers import Adam,RMSprop
```

```python
# Tensorflow version
tf.__version__
```

```python
# Data Augumentation
gen=ImageDataGenerator(
            rescale=1./255,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True
            )
```

```python
training=gen.flow_from_directory(r'D:\Gowsik\Artificial Intelligence Engineering\Deep Learning\Practice Datasets\1577957291_deeplearningwithkerasandtensorflow\data\train',class_mode='binary',
    batch_size=32,target_size=(64,64))
```

```python
gen=ImageDataGenerator(
            rescale=1./255)
```

```python
testing=gen.flow_from_directory(r'D:\Gowsik\Artificial Intelligence Engineering\Deep Learning\Practice Datasets\1577957291_deeplearningwithkerasandtensorflow\data\test',class_mode='binary',
    batch_size=32,target_size=(64,64))
```

```python
model=Sequential()
```

```python
# Feature detector and Maxpooling 1
model.add(Conv2D(filters=32,kernel_size=5,activation='relu',input_shape=[64,64,3]))
model.add(MaxPool2D(pool_size=2,strides=2))
```

```python
# Feature detector and Maxpooling 2
model.add(Conv2D(filters=64,kernel_size=5,activation='relu'))
model.add(MaxPool2D(pool_size=2,strides=2))
```

```python
model.add(Flatten())
```

```python
# Full connection
model.add(Dense(units=32,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(units=1,activation='sigmoid'))
model.summary()
```

```python
# Compiling and Training
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
history=model.fit(x=training,validation_data=testing,epochs=300,verbose=1)
```

```python
# Evaluation
result=model.evaluate(testing,verbose=0)
result
```
